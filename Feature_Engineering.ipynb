# 1. Feature Transformation

## 1.1. Feature Encoding

### 1.1.1. Label Encoding

```python
df1 = df.copy()
df1
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
label_cols = ['Response','Coverage','Education','RenewOfferType','VehicleSize']

for col in label_cols:
    df1[col] = encoder.fit_transform(df1[col])

df1
```

### 1.1.2. Ordinal Encoding

```python
from sklearn.preprocessing import OrdinalEncoder

encoder = OrdinalEncoder()
```

```python
ordinal_cols = ['Coverage','Education','RenewOfferType','VehicleSize']

for col in ordinal_cols:
    df1[col] = encoder.fit_transform(df1[[col]])
df1
```

### 1.1.3. One-Hot Encoding

```python
from sklearn.preprocessing import OneHotEncoder

obj = OneHotEncoder(sparse_output = False, dtype = int)
```

```python
OneHotEncoding_cols = ['State','EmploymentStatus','Gender',
                            'SalesChannel','PolicyType','Policy','VehicleClass','LocationCode','MaritalStatus']
new_col = obj.fit_transform(df1[OneHotEncoding_cols])

encoded_df1 = pd.DataFrame(new_col, columns = obj.get_feature_names_out(OneHotEncoding_cols))

encoded_df1
```

```python
df1 = pd.concat([df1,encoded_df1],axis = 1)

df1
```

```python
df1 = df1.drop(['Customer','State','EmploymentStatus','Gender',
            'SalesChannel','PolicyType','Policy','VehicleClass','LocationCode','MaritalStatus'],axis = 1)
```

## 2.2. Chi-Square Test

```python
from scipy.stats import chi2_contingency

# We'll use df here; change to df1 if needed
categorical_columns = ['RenewOfferType', 'VehicleSize']

# Loop through each categorical column and perform chi-square test with 'Response'
for col in categorical_columns:
    contingency_table = pd.crosstab(df[col], df['Response'])
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    print(f"Chi-square test for '{col}': p-value = {p:.4f}")
```

## 3.1. Over Sampling:

```python
df1['Response'].value_counts().plot(kind='bar', title='Class Distribution', color=['blue', 'orange'])
plt.xlabel('Response')
plt.ylabel('Count')
plt.show()
```

### 3.1.1 Random Over Sampler

```python
from imblearn.over_sampling import RandomOverSampler,SMOTE

# Split into features (X) and target (y)
X = df1.drop('Response', axis=1)
y = df1[['Response']]

# Apply RandomOverSampler
ros = RandomOverSampler(sampling_strategy='not majority')
X_ros, y_ros = ros.fit_resample(X, y)

# Check balanced target counts
print(y_ros.value_counts())
```

### 3.1.2. SMOTE

```python
# Step 1: Drop target and datetime column
X = df1.drop(['Response', 'EffectiveToDate'], axis=1)
y = df1['Response']  # Keep it as a Series

# Step 2: Convert categorical variables to numeric
X_encoded = pd.get_dummies(X, drop_first=True)

# Step 3: Apply SMOTE
smote = SMOTE(sampling_strategy=0.5)
X_smote, y_smote = smote.fit_resample(X_encoded, y)

# Step 4: Check the result
print(y_smote.value_counts())
```

## 3.2. Under Sampling:

### 3.2.1. Random under sampler

```python
from imblearn.under_sampling import RandomUnderSampler

# Features & target
X = df1.drop(['Response', 'EffectiveToDate'], axis=1)
y = df1['Response']

# One-hot encode categorical variables
X_encoded = pd.get_dummies(X, drop_first=True)

# Apply Random UnderSampler
rus = RandomUnderSampler(sampling_strategy='not minority')
X_rus, y_rus = rus.fit_resample(X_encoded, y)

# Check result
print(y_rus.value_counts())
```

### 3.2.2. Edited Nearest Neighbours:

```python
from imblearn.under_sampling import EditedNearestNeighbours

# Apply ENN
enn = EditedNearestNeighbours(n_neighbors=3)
X_enn, y_enn = enn.fit_resample(X_encoded, y)

# Check result
print(y_enn.value_counts())
```
